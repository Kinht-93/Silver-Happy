```python
import pandas as pd
import numpy as np
import os

os.makedirs('data', exist_ok=True)
os.makedirs('outputs', exist_ok=True)
print('Dossiers créés')
```

```python
ages = ['60-69', '70-79', '80+']
regions = ['Île-de-France', 'PACA', 'Auvergne-Rhône-Alpes', 'Occitanie', 'Bretagne']
subs = ['Basique', 'Premium']
mobility = ['Faible', 'Moyenne', 'Élevée']
categories = ['Santé', 'Loisirs', 'Accompagnement', 'Administratif']
status = ['Confirmé', 'Annulé', 'Terminé']

services = {
    'Santé': ['Suivi médical', 'Bien-être', 'Kiné', 'Nutrition'],
    'Loisirs': ['Mémoire', 'Sortie', 'Créatif', 'Jeux'],
    'Accompagnement': ['Courses', 'Visite', 'Support', 'Transport'],
    'Administratif': ['Démarches', 'Courrier', 'Numérique', 'Fiscalité']
}

print('Références définies')
```

```python
data = []

for i in range(1000):
    age = np.random.choice(ages, p=[0.35, 0.33, 0.32])
    region = np.random.choice(regions)
    sub = np.random.choice(subs, p=[0.62, 0.38])
    months = np.random.randint(1, 61)
    mob = np.random.choice(mobility, p=[0.3, 0.5, 0.2])
    
    if age == '80+':
        cat_probs = [0.42, 0.34, 0.16, 0.08]
    elif age == '70-79':
        cat_probs = [0.2, 0.26, 0.38, 0.16]
    else:
        cat_probs = [0.1, 0.16, 0.46, 0.28]
    
    cat = np.random.choice(categories, p=cat_probs)
    svc = np.random.choice(services[cat])
    
    data.append({
        'id': f'C{i+1:04d}',
        'age': age,
        'region': region,
        'subscription': sub,
        'months': months,
        'mobility': mob,
        'category': cat,
        'service': svc,
        'status': np.random.choice(status, p=[0.7, 0.15, 0.15])
    })

df = pd.DataFrame(data)
df.to_csv('data/dataset.csv', index=False)
print(f'Dataset généré: {len(df)} lignes')
print(df.head())
```
