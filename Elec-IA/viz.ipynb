```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

os.makedirs('outputs', exist_ok=True)
df = pd.read_csv('data/dataset.csv')
print(f'Dataset chargé: {df.shape}')
print(df.head())
```

```python
print(f'Valeurs manquantes: {df.isnull().sum().sum()}')
print(f'Doublons: {df.duplicated().sum()}')
print(f'\nDistribution catégories:\n{df["category"].value_counts()}')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 4), facecolor='white')
fig.suptitle('Vue d\'ensemble clients', fontsize=14, fontweight='bold')

df['age'].value_counts().sort_index().plot(kind='bar', ax=ax[0])
ax[0].set_title('Clients par âge')

df['region'].value_counts().head(5).plot(kind='bar', ax=ax[1])
ax[1].set_title('Top régions')

df['subscription'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=ax[2])
ax[2].set_title('Abonnements')
ax[2].set_ylabel('')

plt.tight_layout()
plt.savefig('outputs/clients.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print('Sauvegardé: outputs/clients.png')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4), facecolor='white')
fig.suptitle('Vue d\'ensemble services', fontsize=14, fontweight='bold')

df['category'].value_counts().plot(kind='bar', ax=ax[0])
ax[0].set_title('Services par catégorie')

df['status'].value_counts().plot(kind='bar', ax=ax[1])
ax[1].set_title('Statuts')

plt.tight_layout()
plt.savefig('outputs/services.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print('Sauvegardé: outputs/services.png')
```
